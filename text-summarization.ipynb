```python
# Collect data by using web scaping
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
    AppleWebKit/537.36(KHTML, like Gecko) Chrome/110.0.5481.178 Safari/537.36"
}
url = "https://www.health.harvard.edu/blog/waiting-for-motivation-to-strike-try-rethinking-that-202201042663"
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, "html.parser")

paragraphs = []
for para in soup.find_all("p"):
    paragraph = para.get_text()
    paragraphs += [paragraph]
    
text = str("\n\n".join(paragraphs))
print(text)
```

```python
# Convert text to sentences and words
import nltk
from nltk import sent_tokenize, word_tokenize

sent_list = sent_tokenize(text)
word_list = word_tokenize(text)
```

```python
# Remove stop words and find weighted frequency of occurence
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

word_frequencies = {}
for word in word_list:
    if word not in stop_words:
        if word not in word_frequencies.keys():
            word_frequencies[word] =1
        else:
            word_frequencies[word] +=1

maximum_frequency = max(word_frequencies.values())

for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequency)
```

```python
# Calculate sentence scores
sentence_scores = {}
for sent in sent_list:
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 25:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
```

```python
# Use heapq library and call its nlargest function to get summary
import heapq

summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)
summary = ' '.join(summary_sentences)
print(summary)
```
